```python
from sklearn.datasets import load_iris
import pandas as pd

import matplotlib.pyplot as plt
# Load the iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Define species names
species_names = iris.target_names

# Scatter plot for Sepal Length vs Sepal Width
plt.figure(figsize=(10, 6))
for species in range(len(species_names)):
    subset = iris_df[iris_df['species'] == species]
    plt.scatter(subset['sepal length (cm)'], subset['sepal width (cm)'], label=species_names[species])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')
plt.legend()
plt.show()

# Scatter plot for Petal Length vs Petal Width
plt.figure(figsize=(10, 6))
for species in range(len(species_names)):
    subset = iris_df[iris_df['species'] == species]
    plt.scatter(subset['petal length (cm)'], subset['petal width (cm)'], label=species_names[species])
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs Petal Width')
plt.legend()
plt.show()

# Histogram for Sepal Length
plt.figure(figsize=(10, 6))
plt.hist(iris_df['sepal length (cm)'], bins=20, color='blue', edgecolor='black')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()

# Histogram for Petal Length
plt.figure(figsize=(10, 6))
plt.hist(iris_df['petal length (cm)'], bins=20, color='green', edgecolor='black')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')
plt.show()
```

```python
import seaborn as sns

# Pairplot of the Iris dataset
sns.pairplot(iris_df, hue='species', palette='bright')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()

# Heatmap of the correlation matrix
plt.figure(figsize=(10, 6))
correlation_matrix = iris_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
# Load the tips dataset from seaborn
tips = sns.load_dataset('tips')

```

```python
# Boxplot of total bill by day
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set1')
plt.title('Boxplot of Total Bill by Day')
plt.show()

# Violin plot of total bill by day
plt.figure(figsize=(10, 6))
sns.violinplot(x='day', y='total_bill', data=tips, palette='Set2')
plt.title('Violin Plot of Total Bill by Day')
plt.show()
```
